```python
import gdelt
import pandas as pd
import numpy as np
```

```python
gd1 = gdelt.gdelt(version=1)

# pull single day, gkg table
results= gd1.Search('2016 Nov 01',table='gkg')
print(len(results))
```

```python
results
```

```python
def parse_theme(themes):
    themes_parsed = []
    themes = themes[~themes.isnull()]
    for event in themes:
        themes_parsed.append(event.lower().replace('_', ' ').split(';'))
    return themes_parsed
        
```

```python
themes = parse_theme(results['THEMES'])
```

```python
themes
```

```python
def count_mentions(themes):
    count = 0
    for i in themes:
        for j in i:
            if 'coffee' in j:
                count += 1
    return count
```
